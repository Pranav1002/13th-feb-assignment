```python
#2) Write a python program to print Python Exception Hierarchy.

def exception_hierarchy(className, ind=0):
    print(' ' * ind + className.__name__)
    for i in className.__subclasses__():
        exception_hierarchy(i,ind+2)
        
exception_hierarchy(Exception)
```

```python
#3) What errors are defined in the ArithmeticError class? Explain any two with an example.
'''
---> Arithmetic Error is a super class of the errors generated by arithmatic operations.
---> Below methods are defined in ArithmeticError class.

1) FloatingPointError
2) OverflowError
3) ZeroDivisionError
4) FPE ( subclass of FloatingPointError)
5) AssertionError

'''

# Example of ZeroDivisionError

try:
    a=10/0
except ZeroDivisionError as e:
    print(e)
# Example of OverflowError
try:
    a=10/3
    print(a)
except FloatingPointError as f:
    print(f)
```

```python
#4) Why LookupError class is used? Explain with an example KeyError and IndexError.
'''
---> LookupError class is the base class for errors that occur when specified key is not found on given index.
---> There ar two sub classes of LookupError.
 1) KeyError
 2) IndexError
'''

# Example of KeyError
try:
    d = {'a':1,'b':2,'c':3}
    print(d['f'])
except KeyError as e:
    print(e)

try:
    c = [1,2,3,4]
    print(c[4])
except IndexError as e:
    print(e)
```
